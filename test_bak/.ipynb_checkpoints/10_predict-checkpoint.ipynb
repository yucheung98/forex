```python
import pandas as pd
import numpy as np
from keras import models
from keras import layers
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.core import Dense,Activation,Dropout
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
train_data = pd.read_csv('data/train_data.csv')
test_data = pd.read_csv('data/test_data.csv')

train_data.drop(['Unnamed: 0'], axis=1, inplace=True)
test_data.drop(['Unnamed: 0'], axis=1, inplace=True)

print(train_data.head(1))
print(test_data.head(1))
print(train_data.columns.to_list())
```

```python
feature = ['team_score', 'n', 'player_score', 'home_ad']  #影响因素
label = ['胜负'] #标签一个，即需要进行预测的值
```

```python
x_train = train_data[feature].values #特征数据
y_train = train_data[label].values #标签数据
print(x_train.shape)
print(y_train.shape)
```

```python
test_data_X = test_data[feature].values
test_data_Y = test_data[label].values
print(test_data_X.shape)
print(test_data_Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=520)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
#隐藏层维数的确定
low = (4+1)**0.5+1
print(low)
high = (4+1)**0.5+10
print(high)
```

```python
model = Sequential()
model.add(Dense(4, activation='relu',input_shape=(X_train.shape[1],)))
model.add(Dense(9, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer = 'rmsprop',# 还可以通过optimizer = optimizers.RMSprop(lr=0.001)来为优化器指定参数.optimizer='rmsprop'
              loss='binary_crossentropy', # 等价于loss = losses.binary_crossentropy
              metrics=['accuracy']) # 等价于metrics = [metircs.binary_accuracy]
model.fit(X_train, Y_train,epochs=80, batch_size=10, verbose=0, validation_data=(X_test,Y_test))
```

```python
model.evaluate(X_test,Y_test)
```

```python
model.predict(test_data_X)
```

```python
results = model.evaluate(test_data_X,test_data_Y)
results
```

```python
res = model.predict(test_data_X).tolist()
print(res)
```

```python
result = []
for i in range(len(res)):
    if res[i][0] > 0.5:
        result.append(1)
    else:
        result.append(0)
```
